```python
import tensorflow as tf
```

```python
# 텐서와 연산이 어떠한 장치에 할당되었는지 출력하기 
tf.debugging.set_log_device_placement(True)

# 텐서 생성
a = tf.constant([
    [1, 1],
    [2, 2]
])

b = tf.constant([
    [5, 6],
    [7, 8]
])

c = tf.matmul(a, b)

print(c)

tf.debugging.set_log_device_placement(False)
```
