【문항2】 파이썬의 내장함수를 오버라이딩하여 아래 기능을 포함하는 복소수 클래스를 정의하시오.

       - 기능
         - 객체를 문자열로 표현하는 함수 (복소수 형태로 출력)
         - 덧셈 연산자
         - 뺄셈 연산자
         - 곱셈 연산자 
         - 길이(복소수의 크기)
         - 비교 연산 (==)
         - 비교 연산 (!=)
         - 절대값 

```python
import math

class ComplexNumber:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  
  def __str__(self):
    return '{}+{}j'.format(self.x, self.y)

  def __add__(self, num):
    new_x = self.x + num.x
    new_y = self.y + num.y
    return ComplexNumber(new_x, new_y)

  def __sub__(self, num):
    new_x = self.x - num.x
    new_y = self.y - num.y
    return ComplexNumber(new_x, new_y)

  def __mul__(self, num):
    if isinstance(num, ComplexNumber):
      new_x = (self.x * num.x) - (self.y * num.y)
      new_y = (self.x * num.y) + (self.y * num.x)
      return ComplexNumber(new_x, new_y)
    else:
      new_x = self.x * num
      new_y = self.y * num
      return ComplexNumber(new_x, new_y)
  
  def __len__(self):
    return self.x **2 + self.y **2

  def __eq__(self, equal):
    if isinstance(equal, ComplexNumber):
      return self.x == equal.x and self.y == equal.y
    else:
      return False
  
  def __ne__(self, equal):
    return not self.__eq__(equal)

  def __abs__(self):
    return math.sqrt(self.x**2 + self.y**2)

a = ComplexNumber(1,2)
b = ComplexNumber(3,2)

print('a :', a)
print('b :', b)
print('a+b :', a+b)
print('a-b :', a-b)
print("a*b :", a*b)
print('len(a) :', len(a))
print('len(b) :', len(b))
print('a == b :', a == b)
print('a != b :', a != b)
print('abs(a) :', a.__abs__())
```
