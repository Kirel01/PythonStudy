```python
import pandas as pd
```

```python
# data : https://www.kaggle.com/datasets/hesh97/titanicdataset-traincsv
train_data = pd.read_csv('./sample_data/train.csv')      # 런타임 이후에는 삭제됨 
train_data.head()
```

#### 새 column 추가 
 - [] 사용하여 추가하기 
 - insert함수 사용하여 원하는 위치에 추가하기 

```python
train_data['Age'] * 2 
```

```python
train_data['Age_double'] = train_data['Age'] * 2 
train_data.head()
```

```python
train_data['Age_tripple'] = train_data['Age'] + train_data['Age_double']
train_data.head()
```

```python
train_data.insert(3, 'Fare10', train_data['Fare'] / 10)
train_data.head()
```

#### column 삭제하기
 - drop 함수 사용하여 삭제
  - 리스트를 사용하여 멀티플 삭제 가능

```python
train_data.drop('Age_tripple', axis=1)      # 열 레벨 
```

```python
train_data.head()     # Age_tripple열 안지워짐. 새로운 복사본을 반환한 것임.
```

```python
train_data.drop(['Age_double', 'Age_tripple'], axis=1)
```

```python
train_data.head()       #역시 원본은 해당 컬럼이 존재함 
```

```python
train_data.drop(['Age_double', 'Age_tripple'], axis=1, inplace=True)  # 연산 자체를 수행함. 어떤 결과를 반환하는 것은 아님.
```
