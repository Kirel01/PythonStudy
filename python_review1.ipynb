```python
# 가변 파라미터
# 전달되는 파라미터의 개수가 고정적이지 않은 경우

def temp(*args):
  return args

temp(10)
temp(10,20)
print(type(temp(10,20,30,40,50)))
(temp(10,20,30,40,50))
```

```python
# 키워드 가변 파라미터
def temp(**kwargs):
  print(type(kwargs))
  print(kwargs)
  for k,v in kwargs.items():
    print(k,v)
    
temp(a=1, b=2, c=3)
```

```python
# 워밍문제
# "GPT(Generative Predefined Transformer)" 문자열을 10번 
# 출력하는 print_gpt 함수를 작성하시오.
def print_gpt():
  for i in range(10):
    print("GPT(Generative Predefined Transformer)")
  
print_gpt()
```

```python
# 워밍문제2
# 입력값 자연수가 홀수인지 짝수인지 판별하는 함수를 작성하시오
# 함수 이름은 odd_even()
# 출력 : 짝수 
#        홀수
#        잘못된 값
def odd_even(x):
  if x % 2 == 0 :
    return "짝수"
  elif x % 2 ==1:
   return "홀수"

odd_even(2)
odd_even(9)
```
