```python
from sklearn.datasets import load_boston  #scikit-learn에서 제공하는 datasets 패키지 호출 
import pandas as pd
```

```python
boston_houseprice_data = load_boston()  # Boston 지역의 주택가격 데이터를 활용하기 위해 데이터 로드 
```

```python
df = pd.DataFrame(
    data = boston_houseprice_data.data,             # 가져온 데이터와 변수(컬럼)명들을 활용하여
    columns = boston_houseprice_data.feature_names  # DataFrame 생성 
)
```

```python
df.head(10)
```

```python
# 나이를 나타내는 컬럼(열) 호출하기 
# 2차원 DataFrame에서 하나의 컬럼만 사용
# => 1차원인 Series로 변환됨 (1차원 구조로 출력됨)
df['AGE']
```

```python
# 253번째, 254번째
# 중앙값 계산 : 253번째 값과 254번째 값의 평균값 
print(
  (
    list(df['AGE'].sort_values())[int(len(df['AGE'])/2-1)] +
    list(df['AGE'].sort_values())[int(len(df['AGE'])/2)]      
  ) /2 
)
```
