2. numpy 이용하여 pi(원주율)값을 출력하시오.
  - 몬테카를로 방법을 이용 => 난수를 발생시켜 어떤 특정한 값(pi)의 근사값을 구하는 것

```python
# 다트를 던졌을 때  전체 원 면적(pi = 1^2)에서 해당 그림처럼 들어갈 수 있는 경우 (pi * 1^2 / 4)의 확률
# (비례식) pi/4 : 1 = 4분원 안에 생성된 점 갯수 : 전체 시도 횟수 (천만번, 1e7)
# pi/4 * 1e7 = 4분원 안에 생성된 점 갯수
# pi/4 = 4분원 안에 생성된 점 갯수 / 1e7
# pi = 4 * 4분원 안에 생성된 점 갯수 / 1e7  
```

```python
total = int(1e7)
points = np.random.rand(total, 2)
print(points)
points.shape
```

```python
# 위 천만개 points 중 어떤 것이 그림의 원호 안에 있는 것인지 알아야 함
# 우리는 원호 안에 있는 것만 필요함
# 원의 방정식에 의해서 x^2 + y^ <= 1 => 원호 안에 있는 것 
```

```python
np.sum(points ** 2, axis = 1) # 각각의 제곱의 합 ,  열 단위로 더함
```

```python
np.sum(points ** 2, axis =1) <1 # 천만개 중에서 원안에 들어가는 것만 구하기 위해서 1보다 작은거만 구함 (False는 1보다 큼)
```

```python
np.sum(np.sum(points ** 2, axis =1) <1) # True 개수를 구함, 전체 천만개 중에서 4분위 호 안에 들어간 points 수
```

```python
# pi = 4 * 4분원 안에 생성된 점 개수 / 1e7
4 * np.sum(np.sum(points ** 2, axis =1) <1) /total # pi의 근사치 구함
```
