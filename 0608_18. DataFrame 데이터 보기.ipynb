#### DataFrame
 - Series가 1차원이라면 DataFrame은 2차원으로 확대된 버전 
 - Excel spreadsheet 생각하면 이해하기 쉬움 
 - 2차원이기 때문에 인덱스가 row, column으로 구성됨 
  - row는 각 개별 데이터를, column은 개별 속성을 의미
 - Data Analysis, Machine Learning에서 data 변형을 위해 가장 많이 사용  


```python
import pandas as pd
```

```python
# data : https://www.kaggle.com/datasets/hesh97/titanicdataset-traincsv
train_data = pd.read_csv('./sample_data/train.csv')      # 런타임 이후에는 삭제됨 
```

#### head, tail 함수
 - 데이터 전체가 아닌, 일부를 간단히 보기 위한 함수 
  - head() : 처음부터 5개만 
  - tail() : 마지막부터 5개만 

```python
train_data.head()
```

```python
train_data.head(n=100)
```

```python
train_data.tail()
```

```python
train_data.tail(10)
```

#### dataframe 데이터 파악
 - shape 속성 (row, column)

 - info 함수 - 데이터 타입, 각 아이템의 개수 등 출력 

```python
train_data.shape            # 891개의 데이터(row), 변수(column)가 12개 
```
