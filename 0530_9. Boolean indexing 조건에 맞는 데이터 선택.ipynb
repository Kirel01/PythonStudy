#### Boolean indexing
  - ndarray 인덱싱 시, bool 리스트를 전달하여 True인 경우만 필터링
  - 예) 짝수인 경우만 찾아보기


```python
import numpy as np
```

```python
x = np.random.randint(1, 100, size=10)
print(x)
```

```python
even_mask = x % 2 == 0
print(even_mask)
```

* bool 리스트를 인덱스로 전달

```python
x[even_mask]
```

```python
x[x % 2 == 0]
```

```python
x[x > 80]
```

#### 다중조건 사용하기
  - 파이썬의 논리 연산자 and, or, not 키워드 사용 불가
  - &, | 사용

```python
x % 2 == 0
x > 80

x[(x % 2 == 0) & (x > 80)]
```

```python
x < 30
x > 50 
x[(x < 30) | (x > 50)]
```

  - 평균 기온이 25도를 넘는 날은 몇 일인가?

```python
len(temp)
```

```python
temp[temp>25.0]
```

```python
len(temp[temp>25.0])
```

```python
np.sum(temp>25.0)
```
