#### ravel, np.ravel()
  - 다차원 배열을 1차원으로 변경
  - 'order' 파라미터
    - 'C' - row 우선 변경
    - 'F' - column 우선 변경

```python
x.ravel()
```

```python
np.ravel(x)
```

```python
np.ravel(x, order= 'C')
```

```python
np.ravel(x, order= 'F')   # 값이 커지는 기준이 열임
```

```python
x = np.arange(15).reshape(3, 5)   # 행을 기준으로 값이 커짐
print(x)
```

```python
temp = x.ravel()
print(temp)
```

```python
print(x)
```

```python
temp[0] = 100
print(temp)
print(x)
```

#### flatten
  - 다차원 배열을 1차원으로 변경
  - ravel과의 차이점 
    - copy를 생성하여 변경함
    - 원본 데이터가 아닌 복사본을 반환
  - 'order' 파라미터
    - 'C' - row 우선 변경
    - 'F' - column 우선 변경

```python
y = np.arange(15).reshape(3,5)
print(y)
```

```python
y.flatten()
```

```python
t2 = y.flatten()
print(t2)
```

```python
t2[0] = 100
print(t2)
print(y)
```

```python
t2 = y.flatten(order='C')
print(t2)
```

```python
t3 = y.flatten(order='F')
print(t3)
```

```python
x = np.arange(30).reshape(2,3,5)
print(x)
```

```python
x.ravel()
```

#### reshape() 함수
  - array의 shape을 다른 차원으로 변경
  - 주의할 점은 reshape한 후의 결과의 전체 원소 개수와 이전 개수가 같아야 가능
  - 사례) 
    - 이미지 데이터 벡터화
      - 이미지는 기본적으로 2차원 혹은 3차원 (RGB)이지만 데이터 트레이닝을 위해 1차원으로 변경하여 사용됨

```python
x = np.arange(36)
print(x)
```

```python
print(x.shape)    # 1차원임
```

```python
print(x.ndim)   # 차원
```

```python
y = x.reshape(6, 6)
print(y)
```

```python
y = x.reshape(-1,6)   # 6행 6열로 바꿈
print(y)
```

```python
y = x.reshape(6,-1)   # 6행 6열로 바꿈
print(y)
```

```python
# x.reshape(6,5)    # 크기가 맞지 않으면 에러남
```

```python
y = x.reshape(6,6)
print(y.shape)
print(y.ndim)
```

```python
y = x.reshape(3, 3, 4)
print(y)
print(y.shape)
print(y.ndim)
```

```python
k = x.reshape(3, 3, -1)   # 나머지 2개가 고정되면 하나는 유추할 수 있으니 -1로 설정해도 됨
print(k)
print(k.shape)
print(k.ndim)
```
