```python
import numpy as np
import pandas as pd
```

```python
en_scores = [42, 69, 56, 41, 57, 48, 65, 49, 65, 58]
ma_scores = [65, 80, 63, 63, 76, 60, 81, 66, 82, 78]
```

```python
scores_df = pd.DataFrame(
    {
        'english' : en_scores,
        'mathmatics' : ma_scores
    },
    index=pd.Index(
        [
            'A', 'B', 'C', 'D', 'E',
            'F', 'G', 'H', 'I', 'J'
        ],
        name='student'
    )
)
```

```python
scores_df
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)

# 산점도 그리기 
# 첫번째 인수가 X축, 두번째 인수가 Y축
ax.scatter(scores_df['english'], scores_df['mathmatics'])
ax.set_xlabel('english')
ax.set_ylabel('mathmatics')

plt.show()
```
