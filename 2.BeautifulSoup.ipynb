```python
from bs4 import BeautifulSoup
```

#### html 문자열 파싱
  - 문자열로 정의된 html 데이터 파싱하기

```python
html = '''
  <html>
    <head>
      <title>BeautifulSoup 테스트</title>
    </head>
    <body>
      <div id='upper' class='test' custom='good'>
        <h3>바드 출시</h3>
        <p>구글, 베타버젼 공식 공개</p>
      </div>
      <div id='lower' class='test' custom='good'>
        <p>바드 바드 바드1</p>
        <p>바드 바드 바드2</p>
        <p>바드 바드 바드3</p>
      </div>
    </body>
  </html>
'''
```

#### find 함수
  - 특정한 html tag를 검색할 때 사용
  - 검색 조건을 명시하여 찾고자 하는 tag를 검색

```python
soup = BeautifulSoup(html)
```

```python
soup.find('h3')
```

```python
soup.find('p')
```

```python
soup.find('div', class_ = 'test')
```

```python
attrs = {'id':'upper', 'class' : 'test'}
soup.find('div', attrs=attrs)
```

#### find_all 함수
  -  find가 조건에 만족하는 하나의 tag만 검색한다면, find_all은 조건에 맞는 모든 tag를 list로 반환함

```python
soup.find_all('div', class_='test')
```

#### get_text 함수
  - tag안의 value를 추출
  - 부모 tag의 경우, 모든 자식 tag의 value를 추출

```python
tag = soup.find('h3')
print(tag)
tag.get_text()
```

```python
tag = soup.find('p')
print(tag)
tag.get_text()
```

```python
tag = soup.find('div', id='upper')
print(tag)
tag.get_text().strip()
```
