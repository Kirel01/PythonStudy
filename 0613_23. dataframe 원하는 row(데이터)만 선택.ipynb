```python
import numpy as np
import pandas as pd
```

```python
# data : https://www.kaggle.com/datasets/hesh97/titanicdataset-traincsv
train_data = pd.read_csv('./sample_data/train.csv')      # 런타임 이후에는 삭제됨 
train_data.head(10)
```

#### dataframe slicing
 - dataframe의 경우 기본적으로 [] 연산자가 **column 선택**에 사용
 - 그러나, slicing은 row 레벨로 지원함 

```python
train_data[7:10]
```

#### row 선택하기
 - Series의 경우 []로 row 선택이 가능하나,
   DataFrame의 경우는 기본적으로 column을 선택하도록 설계됨 
 - .loc, .iloc로 row 선택 가능 
  - loc : 인덱스 자체를 사용 
  - iloc : 0 based index로 사용 
  - 이 두 property는 ,를 사용하여 column 선택도 가능    

```python
train_data.head()
```

```python
train_data.info()
```

```python
train_data.index = np.arange(100, 991)
```

```python
train_data.head()
```

```python
train_data.tail()
```

```python
train_data.loc[986]         # 986번 하나만 Series 형태로 가져옴 
```

```python
train_data.loc[[986, 100, 110, 990]]   # 복수개의 row를 가져오려면 리스트 형태로 실행
```

```python
# train_data.loc[[0, 100, 110, 990]]      # 존재하지 않는 index 사용시 에러발생 
```

```python
train_data.iloc[0]
```

```python
train_data.iloc[[0, 100, 200, 2]]     # 리스트 형태로 실행 가능함 
```

#### row, column 동시에 선택하기 
 - loc, iloc 속성을 이용할 때 콤마를 이용하여 둘 다 명시 가능

```python
train_data.loc[[986, 100, 110, 990], ['Survived', 'Name', 'Sex', 'Age']]  # low, column 순서대로함 
```

```python
train_data.iloc[[101, 110, 200, 102], [1, 4, 5]]  # 존재하는 로우와 컬럼들만 명시해서 요청 
```
