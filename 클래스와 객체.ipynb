#### \_\_init__(self)
  + 생성자, 클래스 인스턴스가 생성될 때 호출됨
  + self인자는 항상 첫번째에 오며 자기 자신을 가리킴
  * 이름이 꼭 self일 필요는 없으나, 관례적으로 self로 써줌

```python
class Person:
  def __init__(self):
    print(self, '생성됨')
    self.name = '이순신'
    self.age = 30

p1 = Person()
p2 = Person()
print(p1)
print(p2)
print(p1.name, p2.name)
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("차무식", 11)
p2 = Person("차무속", 22)
p3 = Person("차무숙", 20)

print(p1.name, p1.age)
print(p2.name, p2.age)
print(p3.name, p3.age)
```

#### **self**
  - \파이썬의 method는 항상 첫번째 인자로 self를 전달
  - self는 현재 해당 메서드가 호출되는 객체 자신을 가리킴
  - Java, C++/C#의 this에 해당
  - 위치는 항상 매 처음의 parameter이며, 이름을 관례적으로 self로 사용

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def sleep(self):
    print(self.name, '은 잠을 잔다.')

a = Person('이순신', 20)
b = Person('이도', 22)

a.sleep()
b.sleep()
```

```python
# Counter
# 멤버함수 추가 - 숫자를 하나 증가(increment)
#                 숫자를 0으로 초기화(reset)
#                 현재 숫자값 출력(print_current_value)

class Counter:
  def __init__(self):
    self.num = 0

  def increment(self):
    self.num += 1
  
  def reset(self):
    self.num = 0

  def print_current_value(self):
    print('현재 값 : ', self.num)

c1 = Counter()
c1.print_current_value()
c1.increment()
c1.increment()
c1.increment()
c1.print_current_value()
c1.reset()
c1.print_current_value()

c2 = Counter()
c2.print_current_value()
c2.increment()
c2.increment()
c2.print_current_value()


```

#### **method type**
  - instance method - 객체로 호출
    - 메서드는 객체 레벨로 호출되기 때문에, 해당 메서드를 호출한 객체에만 영향을 줌
    - Class method(static method)
    - 클래스 메서드의 경우 클래스 레벨로 호출됨
    - 클래스 멤버 변수만 변경 가능함

```python
class Math:

  @staticmethod
  def add(a, b):
    return a + b

  @staticmethod
  def multiply(a, b):
    return a * b

Math.add(10,20)
Math.multiply(10,20)
```

#### **Class inheritance (상속)**
  - 코드를 재사용할 수 있게 됨
  - 상속받고자 하는 대상인 기존 클래스를 Parent, Super, Base class 라고 부름
  - 상속받는 새로운 클래스를 Child, Supf

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def eat(self, food):
    print('{}은 {}를 먹습니다.'.format(self.name, food))

  def sleep(self, minute):
    print('{}은 {}분 동안 잡니다.'.format(self.name, minute))

  def work(self, minute):
    print('{}은 {}분 동안 일합니다.'.format(self.name, minute))

class Student(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

class Employee(Person):
  def __init__(self, name, age):
    self.name = name
    self.age = age

cha = Person('차무식', 50)
cha.eat('chicken')
cha.sleep(30)
cha.work(30)

oh = Student('오씨', 30)
oh.eat('bap')
oh.sleep(40)
oh.work(20)

oh = Employee('박씨', 60)
oh.eat('jam')
oh.sleep(100)
oh.work(50)
```
