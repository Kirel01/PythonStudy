#### 브로드캐스팅
  - Shape이 같은 두 nd array에 대한 연산은 각각의 원소별로 진행
  - 연산되는 두 ndarray가 서로 다른 Shape을 갖는 경우 브로드 캐스팅(Shape을 맞춤) 후 적용

#### 브로드캐스팅 Rule
  - [공식문서](https://numpy.org/doc/stable/_images/broadcasting_2.png)
  - 뒷 차원에서부터 비교하여 Shape가 같거나, 차원 중 값이 1인 것이 존재하면 가능

* Shape이 같은 경우의 연산

```python
x = np.arange(15).reshape(3,5)
y = np.random.rand(15).reshape(3,5)
print(x)
print(y)
```

```python
x + y 
```

* Scalar(상수)와의 연산

```python
x + 2
```

```python
x **2
```

```python
x % 2 == 0
```

* Shape이 다른 경우 연산

```python
a = np.arange(12).reshape(4,3)
b = np.arange(100, 103)
print(a)
print(b)
```

```python
a + b   # b가 stretch되어 연산이 가능해짐
```

```python
a = np.arange(12).reshape(4,3)
b = np.arange(100, 103)   # 3행짜리 벡터
c = np.arange(1000, 1004)   # 4행짜리 벡터

print(a)
print(b)
print(c)

print(a.shape)
print(b.shape)
print(c.shape)
                        # 브로드캐스팅은 뒷 차원에서부터 비교함
                        # (4,3)과 (3, )은 뒷 차원이 같음 
                        # (4, )와 (4, )은 뒷 차원이 다름 
```

```python
# a + c   # shape이 안맞음
```

```python
a = np.arange(12).reshape(4,3)
b = np.arange(100, 103)   # 3행짜리 벡터
c = np.arange(1000, 1004)   # 4행짜리 벡터

d = b.reshape(1,3)    # 1행 3열의 행렬

print(a.shape)
print(b.shape)
print(c.shape)
print(d.shape)
print(d)
```

```python
a + d
```

```python
a + b # b가 streth되어 연산 가능
```
