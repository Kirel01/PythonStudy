#### numpy.linalg.inv
- 역행렬을 구할 때 사용하는 함수
- 모든 차원의 값이 같아야함


```python
x = np.random.rand(3,3)     #정사각 행렬만 가능
print(x)

np.linalg.inv(x)      #역행렬을 구할 수 있는 경우는 정사각 행렬의 경우만 가능
```

```python
x @ np.linalg.inv(x)
```

```python
np.matmul(x, np.linalg.inv(x))
```

```python
x = np.random.rand(3,3,3)
np.linalg.inv(x)
```

```python
x @ np.linalg.inv(x)
```

####**numpy.linalg.solve**
- Ax = B 형태의 선형대수식 솔루션을 제공하는 함수
- 예) x와 y의 합은 25
      2x와 4y의 합은 64

```python
A = np.array([[1,1],[2,4]])
B = np.array([[25],[64]])

x = np.linalg.solve(A,B)

print(x)
```

```python
np.allclose(A@x, B)     # 두 수를 비교하는 함수
```
