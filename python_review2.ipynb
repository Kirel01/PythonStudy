```python
# 함수 사용
def cake1(fruit, snack):
  print("{}이 있습니다.".format(fruit))
  print("{}이 있습니다.".format(snack))

def cake2(fruit, snack):
  print("{}가 있습니다.".format(fruit))
  print("{}이 있습니다.".format(snack))

cake1("수박", "핫덕")
cake2("참외","홈런볼")
```

```python
# 클래스 사용 (init 생성자 사용)
class Cake:
  def __init__(self): #자기가 가리키고 있는 주소 즉 자기 지신
    self.fruit = "수박"
    self.snack = "핫덕"

a = Cake()
b = Cake()
print(a.fruit)
print(b.snack)

a.fruit = "딸기"
print(a.fruit)
```

```python
class Cake:
  def __init__(self, fruit, snack):
    self.fruit = fruit
    self.snack = snack

cake1 = Cake("수박", "핫덕")
cake2 = Cake("참외", "홈런볼")

print(cake1.fruit, cake1.snack)
print(cake2.fruit, cake2.snack)
```

```python
class Cake:
  def __init__(self, fruit, snack, price):
    self.fruit = fruit
    self.snack = snack
    self.price = price

  def sell(self):
    print("이 케익은 {}원에 팔았습니다".format(self.price))

a = Cake("수박", "핫덕", 10000)
a.sell()
a.sell()
```

```python
class Cake:
  def __init__(self, fruit, snack, price):
    self.fruit = fruit
    self.snack = snack
    self.price = price
    self.sales = 0

  def sell(self):
    self.sales += self.price
    print("이 케익은 {}원에 팔았습니다".format(self.price))

  def gain(self):
    print("케익은 총 {}원 팔았습니다.".format(self.sales))

a = Cake("수박", "핫덕", 10000)
a.sell()
a.sell()
a.sell()
a.gain()

b = Cake("포도", "꼬깔콘", 5000)
b.sell()
b.sell()
b.sell()
b.gain()
```

```python
# 클래스 상속
# 상속하려 하는 클래스는 = parent
# 상속받으려 하는 새로운 클래스는 = child 

# 부모 클래스
class Car:
  def __init__(self, name, km):
    self.name = name
    self.km = km

  def paint(self, color):
    print("{}차 색상은 {}입니다.".format(self.name, color))

  def order(self, day):
    print("{}차는 출차 기간이 {}일 입니다.".format(self.name, day))

# 자식 클래스
class SuperCar(Car):
  def __init__(self, name, km):
    self.name = name
    self.km = km

a = Car("현기", "500")
a.paint("황금색")
a.order(90)

b = SuperCar("창세기", "300")
b.paint("검정색")
b.order(100)
```
