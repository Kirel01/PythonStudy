```python
x = np.arange(15).reshape(3,5)
y = np.random.rand(15).reshape(3,5)   # 행렬 크기가 다르면 연산 에러 발생함 (shape이 같아야 함) ex) reshape(4,4)
print(x)
print(y)
```

#### 통계 함수
  - 평균, 분산, 중앙, 최대 , 최소값 등 통계 관련된 함수가 내장됨

```python
print(y)
```

```python
np.mean(y)  # 평균
```

```python
y.mean()  # 평균
```

```python
np.max(y)
```

```python
np.argmax(y)    # 가장 큰 값의 인덱스를 가져옴
```

```python
np.var(y), np.median(y), np.std(y)    # 분산, 중앙, 표준편차 계산
```

#### 집계 함수
  - 합(sum), 누적 합계(cumsum) 등등 계산 가능

```python
y
```

```python
sum(y)
```

```python
np.sum(y, axis=0)   # 축을 따라 합을 구함 (행)
```

```python
np.sum(y, axis=1)   # 축을 따라 합을 구함 (열)
```

```python
np.sum(y, axis=None)    # None 이면 전체의 합을 구함
```

```python
np.cumsum(y) # 누적 합계 
```

#### any, all 함수 
  - any : 특정 조건을 만족하는 것이 하나라도 있으면 True, 아니면 False
  - all : 모든 원소가 특정 조건을 만족한다면 True, 아니면 False

```python
z = np.random.randn(10)
print(z)
```

```python
z > 0 
```

```python
np.any(z > 0)
```

```python
np.all(z > 0)   # loop가 필요 없음
```

```python
np.all(z !=0)
```

#### where 함수
  - 조건에 따라 선별적으로 값을 선택 가능
  - 예) 음수인 경우는 0, 나머지는 그대로 값을 쓰는 경우 

```python
z = np.random.randn(10)
print(z)
```

```python
np.where(z > 0, z, 0)
```
