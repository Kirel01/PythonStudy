```python
import torch
```

```python
data = [
    [1, 2],
    [3, 4]
]

x = torch.tensor(data)
print(x.is_cuda)

#GPU로 옮기기
x = x.cuda()
print(x.is_cuda)

#CPU로 옮기기
x = x.cpu()
print(x.is_cuda)
```

```python
tensor = torch.rand(3, 4)
tensor = tensor.cuda()

print(tensor)
print(f"Shape : {tensor.shape}")
print(f"Data type: {tensor.dtype}")
print(f"Device: {tensor.device}")
```

```python
# GPU 장치의 텐서
a = torch.tensor([
    [1, 1],
    [2, 2]
]).cuda()

# CPU 장치의 텐서
b = torch.tensor([
    [5, 6],
    [7, 8]
])

#print(torch.matmul(a, b))  # 오류 발생
print(torch.matmul(a.cpu(), b))
```
