```python
import numpy as np
```

#### np.array 함수로 생성하기

```python
x = np.array([1,2,3,4])
print(x)

y = np.array([[2,3,4],[1,2,5]])   # 2행 3열의 행렬 생성
print(y)

print(type(y))
```

#### np.arrange 함수로 생성하기

```python
np.arange(10)
```

```python
np.arange(1,10) # python의 range()와 비슷한 기능
```

```python
np.arange(1,10,2)
```

```python
# 5에서 100까지 간격 5인 ndarray 생성
np.arange(5, 101, 5)
```

#### np.ones, np.zeros로 ndarray 생성하기

```python
np.ones((4,5))
```

```python
np.ones((2,3,4))    # 3행 4열 행렬이 2개가 생성됨 
```

```python
np.zeros((2,3,6,6))
```

#### np.empty.np.full로 ndarray 생성하기

```python
np.empty((3,4))   # 특정값으로 초기화 되지 않음, 임의의 값으로 채워짐
```

```python
np.full((3,4,2),7)
```

#### np.eye로 ndarray 생성하기
  - 단위 행렬 생성

```python
np.eye(3)   # 3의 3월 행렬을 만들며 대각선은 1로 생성함
```

```python
np.eye(5)
```

#### np.linspace로  ndarray 생성하기

```python
np.linspace(1, 10, 3)     # 1부터 10 사이의 3개의 수를 생성하는데, 그 간격이 일정하게 만들어라
```

```python
np.linspace(1,10,4)
```

```python
np.linspace(1,10,5)
```

#### reshape 함수 사용
  - ndarray의 형태, 차원을 바꾸기 위해 사용

```python
x = np.arange(1,16)
print(x) 

x.shape   # 튜플로 표시됨, 1차원이니까 1개로만 표시됨 

x.reshape(3,5)
x.reshape(5,3)
x.reshape(5,3,1)

# x.reshape(5,4) #숫자가 최소한 20개는 있어야 함, 그러나 현재 x는 15개 밖에 없기 때문에 에러가 발생했음
```
