```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

####%matplotlib inline
- 이미지가 이 문서안에 그려지게 설정

```python
x = np.linspace(0,10,11)
y = x ** 2 + x + 2 + np.random.randn(11)
print(x)
print(y)
```

####**그래프 출력하기**
- plot(x,y) - 선 그래프, scatter() - 점 그래프, hist() - 히스토그램 등 사용
- 함수의 파라미터 혹은 plt의 다른 함수
- 기본적으로 x,y에 해당하는 값이 필요


```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

####**그래프의 주석 추가**
- x, y 축 및 타이틀 
- grid 추가
- x, y 축 범위 추가

```python
plt.xlabel('X values')    #X 레이블 추가
plt.ylabel('Y values')    #Y 레이블 추가
plt.title('x-y relation') #제목추가
plt.grid(True)            #그리드 추가

plt.plot(x,y)
```

```python
plt.xlabel('X values')    #X 레이블 추가
plt.ylabel('Y values')    #Y 레이블 추가
plt.title('x-y relation') #제목추가
plt.grid(True)            #그리드 추가

plt.xlim(0,20)
plt.ylim(0,200)

plt.plot(x,y)
```

####**plot()의 parameters**
- 그래프 형태에 대한 제어 기능
- [plot 함수 다큐먼트] (https://matplotlib.org/stable/api/pyplot_summary.html#module-matplotlib.pyplot)

```python
plt.plot(x,y, '#ff7f0e', )
```

```python
plt.plot(x,y,'.')
```

```python
plt.plot(x,y,'g^')
```

```python
plt.plot(x,y,'m:', linewidth=13)
```

```python
plt.plot(x,y,color='black', linestyle='--', marker='^', markersize='10'
      , markerfacecolor='blue')
```

####**subplot으로 여러 그래프 출력하기**
- subplot함수로 구획을 구별하여 각각의 subplot에 그래프 출력

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'g')

plt.subplot(2,2,3)
plt.plot(y,x,'k')

plt.subplot(2,2,4)
plt.plot(x, np.exp(x),'b')
```

####**hist함수 사용**
- histogram 생성
- bins로 histogram bar 개수 설정


```python
data = np.random.randint(1,100, size=200)
print(data)
```

```python
#bins는 막대그래프 갯수, alpha는 막대그래프의 투명도
plt.hist(data,bins=20, alpha=0.3)
plt.xlabel('x-value')
plt.ylabel('amount')
plt.grid(True)
```
